```python
# Splitting data in input(x) and output(Y)
x=heart_data.drop('CHD', axis=1)
x.shape
```

```python
y=heart_data['CHD']
y.shape
```

```python
# SMOTE (synthetic Minority Oversampling)
# Importing SMOTE module
from imblearn.over_sampling import SMOTE
```

```python
os=SMOTE(random_state=0)
x_os, y_os=os.fit_resample(x,y)
```

```python
x_os.shape,y_os.shape
```

```python
# Now the values are equaly distributed
y_os.value_counts()
```

```python
y_os.value_counts().plot(kind='bar')
plt.show()
```

```python
sns.countplot(x=y_os)
plt.show()
```

```python
heart_data_os=pd.DataFrame(x_os)
heart_data_os['CHD']=y_os
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=heart_data_os,x='totChol',y='sysBP',hue='CHD')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting for training and testing purpose
x_train,x_test,y_train,y_test=train_test_split(x_os,y_os,test_size=0.3,random_state=0)
```

```python
# Importing module for scalling data
from sklearn.preprocessing import StandardScaler
```

```python
# Scalling train Features for optimal performance
sc_train= StandardScaler().fit(x_train)
x_train_sc=sc_train.transform(x_train)
```

```python
np.set_printoptions(precision=3)
print(x_train_sc[0:5,:])
```

## Training of Model

```python
# Importing the Module for the model 
from sklearn.linear_model import LogisticRegression
```

```python
# Creating the instance of the Model
Model= LogisticRegression(solver='liblinear')
```

```python
# Training the model by using training data
Model.fit(x_train_sc,y_train)
```

```python
# Scalling the test data for optimal model performance
sc_test= StandardScaler().fit(x_test)
x_test_sc=sc_test.transform(x_test)
```

```python
# Testing the performnace of the model with test Dataset 
result=Model.score(x_test_sc,y_test)
result*100
```

```python
# Testing the performnace of the model with training data set
result=Model.score(x_train_sc,y_train)
result*100
```

```python
# Graphical representation of correlation between the features and target 
c=['red','green','yellow','purple','blue','orange','cyan']
x.corrwith(heart_data['CHD']).plot.bar(figsize=(20,10),
                                                title='Correlation with CHD',
                                                fontsize=15,rot=45,grid=True,color=c)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Generating the instance of KFold and logistic regression model
kfold=KFold(n_splits=10, random_state=7,shuffle=True)
Model2=LogisticRegression(solver='liblinear')
```

```python
# Training logistic regression model with KFold 
result=cross_val_score(Model2,x_train_sc,y_train,cv=kfold)
result*100
```

```python
# Generating the average of scoring accuracy
result.mean()*100, result.std()*100
```

```python
# Logistic regression and cv with log_loss metric
kfold=KFold(n_splits=10, random_state=7,shuffle=True)
Model3=LogisticRegression(solver='liblinear')
result=cross_val_score(Model3,x_train_sc,y_train,cv=kfold, scoring='neg_log_loss')
result.mean(),result.std()
```

```python
# Logistic regression and cv with AUC metric
kfold=KFold(n_splits=10, random_state=7,shuffle=True)
Model4=LogisticRegression(solver='liblinear')
result=cross_val_score(Model4,x_train_sc,y_train,cv=kfold, scoring='roc_auc')
print("AUC : %.3f  (%.3f)" %(result.mean(),result.std()))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
pred_y=Model.predict(x_test_sc)
matrix=confusion_matrix(y_test,pred_y)
sns.heatmap(matrix/np.sum(matrix), fmt='.2%', annot=True)
```

```python
report=classification_report(y_test,pred_y)
print(report)
```
